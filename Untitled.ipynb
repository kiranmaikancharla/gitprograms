```python
# Read a number as a input and output should be even sum
# Input : 10
# Output : 30(2+4+6+8+10)

n = int(input("Enter the num:"))
i = 0 
sum = 0
while i <= n:
    if i % 2 is 0 :
        sum = sum + i
    i = i + 1
print("The sum is",sum)
```
